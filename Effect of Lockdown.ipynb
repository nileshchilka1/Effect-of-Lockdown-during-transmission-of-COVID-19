```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',None)
%matplotlib inline
```

```python
usa_df = pd.read_csv('us_counties_covid19_daily.csv')
```

```python
usa_df
```

```python
USA_df = pd.DataFrame([usa_df.groupby('date')['cases'].sum() , usa_df.groupby('date')['deaths'].sum() ]).T
```

```python
USA_df.reset_index(inplace=True)
```

```python
USA_df['date'] = np.arange(1,len(USA_df)+1)
```

```python
USA_df.set_index('date',inplace=True)
```

```python
USA_df
```

```python
India_df = pd.read_csv('nation_level_daily.csv')
```

```python
India_df = India_df.iloc[:,:3]
```

```python
India_df
```

```python
India_df['date'] = np.arange(1,len(India_df)+1)
```

```python
India_df.set_index('date',inplace=True)
```

```python
India_df
```

```python
USA_df = USA_df[:111]
```

```python
plt.style.use('seaborn-colorblind')
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
plt.plot(India_df.index[:55],India_df['totalconfirmed'][:55],label='India - ConfirmedCases')
plt.plot(USA_df.index[:55],USA_df['cases'][:55],label='USA - ConfirmedCases')


plt.plot(India_df.index[:55],India_df['totaldeceased'][:55],label='India - Deaths')
plt.plot(USA_df.index[:55],USA_df['deaths'][:55],label='USA - Deaths')
plt.xlabel('Number of Days')
plt.ylabel('Count')
plt.title('First 55 Days of India and USA without Lockdown')
plt.legend()

plt.subplot(1,2,2)
plt.plot(India_df.index[55:],India_df['totalconfirmed'][55:],label='India - ConfirmedCases')
plt.plot(USA_df.index[55:],USA_df['cases'][55:],label='USA - ConfirmedCases')


plt.plot(India_df.index[55:],India_df['totaldeceased'][55:],label='India - Deaths')
plt.plot(USA_df.index[55:],USA_df['deaths'][55:],label='USA - Deaths')
plt.xlabel('Number of Days')
plt.ylabel('Count')
plt.title('Next 56 Days of India(Complete Lockdown) and USA(Partially Lockdown)')
plt.legend()

plt.show()
```
